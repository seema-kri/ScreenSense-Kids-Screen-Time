# week 3

# Age Group

```python
dataset["Age_Group"] = pd.cut(
    dataset["Age"],
    bins=[7,10,13,16,18],
    labels=["8-10","11-13","14-16","17-18"]
)
```

# Screen Time Category

```python
dataset["Screen_Category"] = pd.cut(
    dataset["Avg_Daily_Screen_Time_hr"],
    bins=[0,3,5,7,15],
    labels=["Low","Moderate","High","Extreme"]
)
```

# Device Type Category

```python
dataset["Device_Type"] = dataset["Primary_Device"].replace({
    "Smartphone": "Portable",
    "Tablet": "Portable",
    "Laptop": "Portable",
    "Tv": "Non-Portable"
})
```

```python
dataset['Device_Type'].value_counts()
```

# Health Issue Count

```python
dataset["Health_Issue_Count"] = dataset["Health_Impacts"].apply(
    lambda x: 0 if x=="No Health Impact" else len(x.split(","))
)
```

# Risk Level

```python
dataset["Risk_Level"] = np.where(
    dataset["Exceeded_Recommended_Limit"]==True,
    "High Risk",
    "Normal"
)
```

# Educational Usage Category

```python
dataset["Usage_Type"] = np.where(
    dataset["Educational_to_Recreational_Ratio"]>0.45,
    "More Educational",
    "More Recreational"
)
```

```python
dataset["Usage_Type"].value_counts()
```

# Screen Time Intensity Score

```python
dataset["Usage_Intensity"] = dataset["Avg_Daily_Screen_Time_hr"] * dataset["Health_Issue_Count"]
```

# Top 5 Groups with Highest Average Screen Time

```python
# Group by Age, Gender, Urban/Rural and calculate mean screen time
peak = dataset.groupby(
    ["Age", "Gender", "Urban_or_Rural"]
)["Avg_Daily_Screen_Time_hr"].mean().reset_index()

# Show top 5 groups with highest average daily screen time
top5_peak = peak.sort_values(
    by="Avg_Daily_Screen_Time_hr", 
    ascending=False
).head(5)

top5_peak
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a label for combined categories
top5_peak['Group'] = top5_peak['Age'].astype(str) + " | " + top5_peak['Gender'] + " | " + top5_peak['Urban_or_Rural']

sns.barplot(
    x="Avg_Daily_Screen_Time_hr",
    y="Group",
    data=top5_peak,
    palette="viridis"
)

plt.xlabel("Average Daily Screen Time (hours)")
plt.ylabel("Age | Gender | Location")
plt.title("Top 5 Groups with Highest Average Daily Screen Time")
plt.show()
```
